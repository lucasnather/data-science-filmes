Este projeto será direcionado para a verificação de Filmes / Estlios que eu gosto como boas notas, para uma melhor filtragem e procura de qual assistir ( pois eu não aguento mais passar 1 hora procurando e dormindo hehehe).

```python
import pandas as pd
```

```python
dados_imdb = pd.read_csv('imdb_top_1000.csv')
dados_imdb.head()
```

# Verificando os tipos de dados da tabela e seus valores nulos

```python
dados_imdb.info()
```

# Preenchendo os Valores não nulos de Meta_Score com 0 , para não perder os filmes listados da coluna IMDB_Ratings

```python
dados_imdb['Meta_score'].fillna(0, inplace=True)
dados_imdb.info()
```

# Removendo colunas que não preciso Link do Poster, Certificados, Estrela 2, Estrela 3, Estrela 4, Não Votos, e bruto

```python
dados_imdb = dados_imdb.drop(['Poster_Link', 'Certificate', 'Star2', 'Star3', 'Star4', 'No_of_Votes', 'Gross'], axis=1)
dados_imdb.head()
```

# Genêros existentes na tabela

```python
dados_genres = dados_imdb['Genre'].value_counts()
dados_genres
```

#  Colocar filmes do IMDB Rating multiplicado por 10 para os dados sairem normalizados na hora de calcular a média

```python
dados_imdb['IMDB_Rating'] *= 10
dados_imdb
```

# Fazendo uma pesquisa verifiquei que o IMDB não leva em consideração as notas do metacritic -> então realizarei a média delas duas no meu projeto para filtrar ainda mais filmes que tem notas boas.

```python
dados_imdb['Mean_Rating'] = (dados_imdb['IMDB_Rating'] + dados_imdb['Meta_score']) / 2
dados_imdb
```

# Verificar filmes com média maior ou igual a 80, ordenando os filmes em ordem descrente

```python
is_ratings_bigger_than_80 = dados_imdb['Mean_Rating'] >= 80.0
dados_imdb = dados_imdb[is_ratings_bigger_than_80]
```

# Removendo Colunas IMDB_Ratings e Meta_score

```python
dados_imdb = dados_imdb.drop(columns=['IMDB_Rating', 'Meta_score'])
dados_imdb
```

# Filtrando Filmes de Ficção Científica em ordenando em ordem Decrescente

```python
sci_fi_filmes = dados_imdb[dados_imdb['Genre'].str.contains('Sci-Fi')].sort_values('Mean_Rating', ascending=False)
sci_fi_filmes
```

# Filtrando Filmes de Terror em ordenando em ordem Decrescente

```python
horror_filmes = dados_imdb[dados_imdb['Genre'].str.contains('Horror')].sort_values('Mean_Rating', ascending=False)
horror_filmes
```

# Filtrando Filmes de Romance em ordenando em ordem Decrescente

```python
romance_films = dados_imdb[dados_imdb['Genre'].str.contains('Romance')].sort_values('Mean_Rating', ascending=False)
romance_films
```
